```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('Train_Data.csv')
test_data = pd.read_csv('Test_Data.csv')
```

```python
train_data['ad'].unique()
```

```python
test_data['ad'].unique()
```

```python
train_data = train_data[train_data.ad.isin(['ad 1', 'ad 2', 'ad 3', 'ad 4', 'ad 5', 'ad 6', 'ad 7', 'ad 8',
       'ad 10', 'ad 71', 'ad 72', 'ad 73', 'ad 74', 'ad 75', 'ad 56',
       'ad 55'])]
```

```python
train_data
```

```python
train_data['ad'].unique()
```

```python
train_data.drop('date', axis=1, inplace=True)
train_data.drop('campaign', axis=1, inplace=True)
test_data.drop('date', axis=1, inplace=True)
test_data.drop('campaign', axis=1, inplace=True)
train_data.drop('ad', axis=1, inplace=True)
test_data.drop('ad', axis=1, inplace=True)
```

```python
train_data = pd.get_dummies(train_data)
test_data = pd.get_dummies(test_data)
train_data.head()
```

```python
X_train = train_data.drop(['revenue'], axis='columns')
y_train = train_data['revenue']
X_test = test_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_train_scaled

X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_tuned = RandomForestRegressor(n_estimators= 300,
                                min_samples_split= 5,
                                min_samples_leaf= 4,
                                max_features= 'auto',
                                max_depth= 20,
                                bootstrap= True)
rf_tuned.fit(X_train_scaled, y_train)
```

```python
preds = rf_tuned.predict(X_test_scaled)
preds = preds.astype('int64')
preds
```

```python
prediction = pd.DataFrame(preds, columns=['revenue']).to_csv('prediction4.csv', index=False)
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
  tf.keras.layers.Dense(10),
  tf.keras.layers.Dense(10),
  tf.keras.layers.Dense(10),
  tf.keras.layers.Dense(10),
  tf.keras.layers.Dense(1),
])

# 2. Compile the model
model.compile( loss = tf.keras.losses.mae,
               optimizer = tf.keras.optimizers.Adam(),
               metrics = tf.keras.metrics.RootMeanSquaredError())

# 3. Fit the model
model.fit(X_train_scaled, y_train, epochs= 10)
```

```python
preds = model.predict(X_test_scaled)
preds = preds.astype('int64')
preds
```

```python
preds[preds < 10] = 0
preds
```

```python
prediction = pd.DataFrame(preds, columns=['revenue']).to_csv('prediction1.csv', index=False)
```
